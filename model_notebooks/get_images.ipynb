```python
from timeit import default_timer as timer

start_time = timer()

file = 'images'
characters = ['Homer Simpson','Marge Simpson', 'Bart Simpson', 'Lisa Simpson']

for character in characters:
  print(f'[INFO] Starting to download images of {character}.')

  for query in character_features[character]:
    num_images = 80

    downloader.download(query, limit=num_images, output_dir=f'{file}/{character}', adult_filter_off=True, force_replace=False, timeout=60, verbose=True)

    print(f'[INFO] {num_images} images were succesfuly installed.')

  print(f'[INFO] All {num_images * len(character_features[character])} images of {character} were succesfuly installed.')

end_time = timer()

print('-'*50)
print(f'Time: {end_time - start_time}')
```

```python
from pathlib import Path
import shutil
import os

for path in Path(f'/content/{file}').glob('*/*/*'):
  os.rename(path, path.parent.parent / f'{"_".join(str(path.parent.stem).split(" "))}_{path.name}')
```

```python
for path in Path(f'/content/{file}').glob('*/'):
  folders = os.listdir(path)
  for folder in folders:
    if os.path.isdir(path/folder):
      shutil.rmtree(path/folder)
```

```python
for path in Path(f'/content/{file}').glob('*/*'):
  os.rename(path, str(path).replace(',',''))
```

```python
import numpy as np
import cv2
import os

def dhash(image, hashSize=8):
	gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
	resized = cv2.resize(gray, (hashSize + 1, hashSize))

	diff = resized[:, 1:] > resized[:, :-1]

	return sum([2 ** i for (i, v) in enumerate(diff.flatten()) if v])

hashes = {}
imagePaths = [str(path) for path in Path(f'/content/{file}').glob('*/*')]

debug_count = 0

for imagePath in imagePaths:
  try:
    image = cv2.imread(imagePath)
    h = dhash(image)

    p = hashes.get(h, [])
    p.append(imagePath)
    hashes[h] = p

  except:
    os.remove(imagePath)

  finally:
    debug_count += 1
    if debug_count % 200 == 0:
      print(f'{debug_count} images checked.')
```

```python
for (h, hashedPaths) in hashes.items():
	if len(hashedPaths) > 1:
		for p in hashedPaths[1:]:
				os.remove(p)
```

```python
for path in Path(f'/content/{file}').glob('*/*.jpg'):
  image = cv2.imread(str(path))
  cv2.imwrite(f'{str(path)[:-4]}.jpeg', image)
  os.remove(path)
```
